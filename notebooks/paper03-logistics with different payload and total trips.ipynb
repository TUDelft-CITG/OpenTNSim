## Logistics with different payload and total trips

## cargo amount=30000ton, total trips vary with different payload each trip
#### waterway: h=4m, distance=300km; one trip =2 * distance.
#### Same vessel:  M8, L* B * T=110* 11.4* 3.5, DWT=3000ton; construction year=1990;
#### assume that the ship maintains a constant speed to water Vs=3,66m/s=13,2km/h 
#### transport stratigies: loaded go( 2500t * 12trips, 2000t * 15 trips, 1500t * 20 trips, 1000t * 30 trips, 500t * 60 trips); unloaded return(0t each time)
## figure output:
######   bar chart, to show our methodology can compare the diesel consumption and emission of different transport strategies.
######  x-axis: 2500t(12 trips) , 2000t (15 trips) , 1500t(20 trips)  ,1000t(30 trips) , 500t(60 trips) .  
###### y-axis: total sailing distance(km),total diesel consumption (kg), total emission(kg) of each transport  strategy. 

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set width of bars
barWidth = 0.15
# set heights of bars 
bars_diesel =  [131.302,161.420,215.864,314.673,686.385] # ton,  
bars_distance = [720.0,900.0,1200.0,1800.0,3600.0]   # km
bars_CO2 = [417.927,514.476,687.848, 1004.517,2174.345] # 1000kg   from 2500 to 500 ton
bars_PM10 = [232,285,381,558,1205]      # kg
bars_NOx  = [582.0,715.9,957.2,1396.3,3036.3] # 10kg


# bars_adft = [350, 300, 260, 210, 170,140]  # cm,  actual draft from 2500 to 0 ton
# bars_P = [643, 607, 614, 554, 740,1494]   # kw,  required power from 2500 to 0 ton
# bars_C02 = [368.70, 351.07, 354.20, 323.92, 415.75,792.21]  # 0.1kg/km,   CO2 emission rate from 2500 to 0 ton 
# bars_PM10 = [203, 194, 196, 180, 229,440]      # 0.01kg/km,     PM10 emission rate from 2500 to 0 ton
# bars_NOx  = [506, 480, 484, 440, 576,1111] # g/km,    NOx emission rate from 2500 to 0 ton
# Set position of bar on X axis,(r1=actual draft(m), r2=required power(kw), r345=emission rate(g/km):CO2,PM10,NOx)
r1 = np.arange(len(bars_diesel))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4] 
# Make the plot
fig, axs = plt.subplots( figsize=(12, 8), constrained_layout=True)
fontsize=20
plt.bar(r1, bars_diesel, color='#aaa662', width=barWidth, edgecolor='white', label='total diesel consumption (ton)')
plt.bar(r2, bars_distance, color='#1d5dec', width=barWidth, edgecolor='white', label='total sailing distance(10km)')
plt.bar(r3, bars_CO2, color='#f97306', width=barWidth, edgecolor='white', label='CO2 total emission(1000kg)')
plt.bar(r4, bars_PM10, color='#7bf2da', width=barWidth, edgecolor='white', label='PM10 total emission(kg)')
plt.bar(r5, bars_NOx, color='#ff69af', width=barWidth, edgecolor='white', label='NOx total emission(10kg)')

plt.plot(r1, bars_diesel, '-o', color='#aaa662')
plt.plot(r2, bars_distance, '-o',color='#77a1b5')
plt.plot(r3, bars_CO2, '-o',color='#f97306')
plt.plot(r4, bars_PM10,'-o', color='#7bf2da')
plt.plot(r5, bars_NOx, '-o',color='#ff69af')
# Add xticks on the middle of the group bars
plt.xlabel('Transport strategies(cargo amount 30000 ton)', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars_NOx))], ['2500 ton (12 trips)','2000 ton (15 trips)' ,
                                                        '1500 ton (20 trips)','1000 ton (30 trips)', '500 ton (60 trips)',])
# Create legend & Show graphic
plt.legend()
plt.show()
```
