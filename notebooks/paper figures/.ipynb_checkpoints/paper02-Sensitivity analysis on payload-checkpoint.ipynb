# Sensitivity analysis on payload 

## Payload: 2500 t, 2000t, 1500t, 1000t, 500t, 0t
####  waterway: h=4m, distance=300km; 
#### Same vessel:  M8, L* B * T=110* 11.4* 3.5, DWT=3000ton; construction year=1990
#### assume that the ship maintains a constant speed to water Vs=3,66m/s=13,2km/h 
## figure output:
###### actrual draft,  required power(kw), emission rate (g/km) 

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set width of bars
barWidth = 0.15
# set heights of bars 
bars_adft = [350, 300, 260, 210, 170,140]  # cm,  actual draft from 2500 to 0 ton
bars_P = [643, 607, 614, 554, 740,1494]   # kw,  required power from 2500 to 0 ton
bars_C02 = [368.70, 351.07, 354.20, 323.92, 415.75,792.21]  # 0.1kg/km,   CO2 emission rate from 2500 to 0 ton 
bars_PM10 = [203, 194, 196, 180, 229,440]      # 0.01kg/km,     PM10 emission rate from 2500 to 0 ton
bars_NOx  = [506, 480, 484, 440, 576,1111] # g/km,    NOx emission rate from 2500 to 0 ton
# Set position of bar on X axis,(r1=actual draft(m), r2=required power(kw), r345=emission rate(g/km):CO2,PM10,NOx)
r1 = np.arange(len(bars_adft))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4] 
# Make the plot
fig, axs = plt.subplots( figsize=(12, 8), constrained_layout=True)
fontsize=20
plt.bar(r1, bars_adft, color='#017371', width=barWidth, edgecolor='white', label='actual draft(cm)')
plt.bar(r2, bars_P, color='#77a1b5', width=barWidth, edgecolor='white', label='required power(kw)')
plt.bar(r3, bars_C02, color='#f97306', width=barWidth, edgecolor='white', label='CO2 emission rate(0.1kg/km)')
plt.bar(r4, bars_PM10, color='#7bf2da', width=barWidth, edgecolor='white', label='PM10 emission rate(0.01kg/km)')
plt.bar(r5, bars_NOx, color='#ff69af', width=barWidth, edgecolor='white', label='NOx emission rate(g/km)')

plt.plot(r1, bars_adft, '-o', color='#017371')
plt.plot(r2, bars_P, '-o',color='#77a1b5')
plt.plot(r3, bars_C02, '-o',color='#f97306')
plt.plot(r4, bars_PM10,'-o', color='#7bf2da')
plt.plot(r5, bars_NOx, '-o',color='#ff69af')
# Add xticks on the middle of the group bars
plt.xlabel('loading level(ton)', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars_NOx))], ['2500', '2000', '1500', '1000', '500','0'])
# Create legend & Show graphic
plt.legend()
plt.show()
```
