# Impact of current on the trip between sea port and inland terminal 

## Trip from sea port to inland terminal and trip from inland terminal to sea port  
####  Waterway: h=4m, distance between sea port and inland terminal=300km; 
#### Same vessel:  M8, L* B * T=110* 11.4* 3.5, DWT=3000ton; actual payload=2500 ton every trip, construcation year 1990
#### assume that the ship maintains a constant speed to land V=3,66m/s=13,2km/h, 
#### ++without current, ship speed to water Vs=3,66m/s=13,2km/h
#### ++with mild current 0.5m/s from inland terminal to sea port:  downstream(inland terminal-->sea port),  Vs= 3,16m/s=11,4km/h; upstream(sea port-->inland terminal), Vs= 4,16m/s=15km/h  
## figure output:
#### barchart  to show the total diesel consumption (kg) and emission (g) of the no current trip,downsteam trip,upstream trip, and the round trip(up+down). compare the up-down round trip with no current round trip.


```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set width of bars
barWidth = 0.15
# set heights of bars 
bars_diesel = [3292,7259,2130,6583,9389 ]   # kg,diesel consumption: no current,downstream,upstream,round trip no current,round trip up+down 
bars_CO2 = [1106.1,2283.4,758.6,2212.1,3042.0]  # 10kg,   
bars_PM10 = [610.5,1251.7,447.9,1220.9,1699.6]  # 0.01kg,     
bars_NOx  = [1517,3161,1048,3034,4209] # 0.1kg,
# Set position of bar on X axis,(r1=diesel consumption, r234=emission rate:CO2,PM10,NOx)
r1 = np.arange(len(bars_diesel))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# Make the plot
fig, axs = plt.subplots( figsize=(12, 8), constrained_layout=True)
fontsize=20

plt.bar(r1, bars_diesel, color='#aaa662', width=barWidth, edgecolor='white', label='Total diesel consumption (kg)')
plt.bar(r2, bars_CO2, color='#f97306', width=barWidth, edgecolor='white', label='Total CO2 emission (10kg)')
plt.bar(r3, bars_PM10, color='#7bf2da', width=barWidth, edgecolor='white', label='Total PM10 emission (0.01kg)')
plt.bar(r4, bars_NOx, color='#ff69af', width=barWidth, edgecolor='white', label='Total NOx emission (kg)')


# plt.plot(r1, bars_diesel, '-o',color='#aaa662')
# plt.plot(r2, bars_CO2, '-o',color='#f97306')
# plt.plot(r3, bars_PM10,'-o', color='#7bf2da')
# plt.plot(r4, bars_NOx, '-o',color='#ff69af')
# Add xticks on the middle of the group bars

plt.xticks([r + barWidth for r in range(len(bars_NOx))], ['single trip - no current, either way', 'single trip - upstream',
                                                 'single trip - downstream', 'round trip - no current', 'round trip - current'])
# Create legend & Show graphic
plt.legend()
plt.show()
```
