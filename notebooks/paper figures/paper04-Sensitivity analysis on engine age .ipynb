## Sensitivity analysis on engine age 

## Engine construction year: 1970,1980,1990,2000,2010,2020
####  waterway: h=4m, distance=300km; 
#### Same vessel:  M8, L* B * T=110* 11.4* 3.5, DWT=3000ton; actual payload=2500 ton
#### assume that the ship maintains a constant speed to water Vs=3,66m/s=13,2km/h 
## figure output:
######  diesel consumption (kg/km),emission rate (g/km)
###### ---to show emission rates vary with engine age, but diesel consumption is the same.

```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# set width of bars
barWidth = 0.15
# set heights of bars 
bars_diesel = [84.2, 80.6,78.8,73.5,71.7,68.1]   # 0.1kg/km,  diesel consumption from 1970 to 2020 
bars_C02 = [270.89,255.85,250.11,232.91,227.54,216.07]  # 0.1kg/km,   CO2 emission rate from 1970 to 2020  
bars_PM10 = [21.75,21.75,14.5,10.88,7.25,0.54]      # g/km,     PM10 emission rate from 1970 to 2020 
bars_NOx  = [366,353,343,319,257,149] # g/km,    NOx emission rate from 1970 to 2020 
# Set position of bar on X axis,(r1=diesel consumption, r234=emission rate:CO2,PM10,NOx)
r1 = np.arange(len(bars_diesel))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# Make the plot
fig, axs = plt.subplots( figsize=(12, 8), constrained_layout=True)
fontsize=20
#plt.bar(r1, bars_adft, color='#017371', width=barWidth, edgecolor='white', label='actual draft(cm)')
plt.bar(r1, bars_diesel, color='#aaa662', width=barWidth, edgecolor='white', label='diesel consumption(0.1kg/km)')
plt.bar(r2, bars_C02, color='#f97306', width=barWidth, edgecolor='white', label='CO2 emission rate(0.1kg/km)')
plt.bar(r3, bars_PM10, color='#7bf2da', width=barWidth, edgecolor='white', label='PM10 emission rate(g/km)')
plt.bar(r4, bars_NOx, color='#ff69af', width=barWidth, edgecolor='white', label='NOx emission rate(g/km)')

#plt.plot(r1, bars_adft, '-o', color='#017371')
plt.plot(r1, bars_diesel, '-o',color='#aaa662')
plt.plot(r2, bars_C02, '-o',color='#f97306')
plt.plot(r3, bars_PM10,'-o', color='#7bf2da')
plt.plot(r4, bars_NOx, '-o',color='#ff69af')
# Add xticks on the middle of the group bars
plt.xlabel('Engine construction year', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars_NOx))], ['1970', '1980', '1990', '2000', '2010','2020'])
# Create legend & Show graphic
plt.legend()
plt.show()
```

```python
diesel =[ 8.42,8.06,7.88,7.35,7.17,6.81]  #kg/km
C02 = [27089,25585,25011,23291,22754,21607] #g/km
PM10 = [21.75,21.75,14.5,10.88,7.25,0.54]#g/km
NOx  = [366,353,343,319,257,149]#g/km
# dataframe sample data
df1 = pd.DataFrame(diesel, columns=['fuel consumption (kg/km)'])
df2 = pd.DataFrame(C02, columns=['CO2 emission (g/km)'])
df3 = pd.DataFrame(PM10, columns=['PM10 emission (g/km)'])
df4 = pd.DataFrame(NOx, columns=['NOx emission (g/km)'])


fig, ((ax1),(ax2),(ax3),(ax4)) = plt.subplots(4,1,figsize=(9.5, 8), constrained_layout=True)
plt.subplot(411)
plt.title("The fuel consumption and emission rates of M8 vessel with different engine ages,h=6m,V=3.5m/s,full load",fontsize=12)
df1.plot(color= '#ad900d',rot=0,fontsize=12,kind='bar', ax=ax1)
plt.xticks([0, 1, 2,3,4,5], ['1970','1980','1990','2000','2010','2020'],rotation=0)  
plt.ylim(6, 10)
plt.subplot(412)
df2.plot( color= '#f97306',rot=0,fontsize=12,kind='bar', ax=ax2)
plt.xticks([0, 1, 2,3,4,5], ['1970','1980','1990','2000','2010','2020'],rotation=0) 
plt.ylim(20000, 30000)
plt.subplot(413)
df3.plot( color= '#7bf2da',rot=0,fontsize=12,kind='bar', ax=ax3)
plt.xticks([0, 1, 2,3,4,5], ['1970','1980','1990','2000','2010','2020'],rotation=0) 
plt.ylim(0, 30)
plt.subplot(414)
df4.plot(color= '#ff69af',rot=0,fontsize=12,kind='bar', ax=ax4)
plt.xticks([0, 1, 2,3,4,5], ['1970','1980','1990','2000','2010','2020'],rotation=0) 
plt.ylim(100, 500)

plt.xlabel("Engine construction years",fontsize=12)

plt.show()
```
